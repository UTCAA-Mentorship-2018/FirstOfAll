```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
import math
import import_ipynb
import Imputing_Missing_Data
```

```python
datafile = 'C:/Projects/DataScienceProject/application_train.csv'
data1 = pd.read_csv(datafile)
df = pd.DataFrame(data1)
data = df.copy()
data_update = Imputing_Missing_Data.drop_duplications(data)
data_update.head()
```

```python
subsets = Imputing_Missing_Data.separate_data(data_update)
```

```python
Missing, NonMissing, SeparatedData = Imputing_Missing_Data.dividing_subset_data(subsets)
```

```python
MissingIndex, NonMissingIndex, MissingArr, NonMissingArr = Imputing_Missing_Data.getting_indexes(Missing, NonMissing)
print('The indexes of the Missing TARGET Means are:')
print(MissingIndex, '\n')
print('The indexes of the Non-Missing TARGET Means are:')
print(NonMissingIndex, '\n')
print('The Non-Missing corresponding TARGET Means are:')
print(NonMissingArr, '\n')
```

```python
CorrespondingIndex = Imputing_Missing_Data.mapping_index(MissingIndex, NonMissingIndex, MissingArr, NonMissingArr)
print('The indexes of the Non-Missing corresponding TARGET Means that needed to imputed to Missing dataset are:')
print(CorrespondingIndex, '\n')
```

```python
UpdatedData = Imputing_Missing_Data.data_imputation(MissingIndex, CorrespondingIndex, SeparatedData)
```

```python
for i in range(len(UpdatedData)):
    MissingData = UpdatedData[i]['Missing']
    pd.DataFrame(data_update).update(MissingData)
    print('The SeparatedData set ' + str(i) + ' has been updated')
```
